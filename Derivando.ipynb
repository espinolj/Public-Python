## Aproximaciones mediante secantes

Dada una funci√≥n $f(x): I \rightarrow \mathbb{R}$, un valor $x_0 \in I$ y un valor $h$ trazamos la recta secante que pasa por los puntos $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$. Conforme el valor de $h$ se va haciendo m√°s peque√±o las rectas secantes se ir√°n aproximando a la recta tangente a la gr√°fica de la funci√≥n en el punto $(x_0,f(x_0))$. 

$\large{\textbf{Ejemplos}}$

$\large{\textbf{Para un valor de $\textit{h}$ determinado}}$

Sea la funci√≥n $f(x)=3x^2+5$, graficar la recta secante, que pasa por los puntos $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$ para los valores $x_0$ y $h$ que se asignen, como ejemplo inicial $x_0=-5$ y $h=8$.

## M√≥dulo 1: Grafica funci√≥n f dada y secante que pasa por $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$, para $x_0$ y $h$ dados

```python
# ***************** CONSIDERACIONES PARA LA DEFINICI√ìN DE FUNCIONES*********************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x), la funcion arcoseno(x) como arcsin(x).
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN"
# Asingar valores para x0 y h despu√©s de la l√≠nea "AQU√ç SE ASIGNAN LOS VALORES x0 y h "
# ***************************************************************************************************

# *******************DEFINICI√ìN DE LA FUNCI√ìN********************************************************
def f(x):
    #Llamada de funciones de numpy
    from numpy import sin, cos, tan, arcsin, arccos, arctan, log, exp
    
    # **************************************************************************** 
    #            AQU√ç SE DEFINE LA FUNCI√ìN    
    f=3*x**2+5
    
    # ****************************************************************************
    
    return f

# ********************************************************************************
#                 AQU√ç SE ASIGNAN LOS VALORES x0 Y h 
x0=-5
h=8

# ********************************************************************************


%matplotlib inline
def Graficasecante(x0,h):
    #De matplotlib se importa pyplot como plt
    import matplotlib.pyplot as plt
    #Se importa numpy como np
    import numpy as np

    
    # Se evalua la funci√≥n en el rango de [x0-h,x0+2h]
    x = np.linspace(x0-h,x0+2*h,101)
    #Se le pasa a la funcion los valores definidos en x.
    y = f(x)

    #Se calcula la recta secante, de la forma y=(y1-y0)/(x1-x0)*(x-x0)+y0
    
    rsec = (f(x0+h)-f(x0))/h*(x-x0)+f(x0)  
    
    #Para el valor h se grafica la recta secante que pasa por (x0,f(x0)) y (x0+h,f(x0+h))
    
    # Se grafica la funcion y la secante.
    plt.title('Secante con h  = '+str(h))
    plt.plot(x,y,'b',x0,f(x0),'om',x0+h,f(x0+h),'om',x,rsec,'r')
    
    # Indicar s√≥lo los valores requeridos en los ejes
    plt.xticks([x0,x0+h])
    plt.yticks([f(x0),f(x0+h),max(y)])
    # Se mueven los ejes coordenados con intersecci√≥n en el origen
    ax = plt.gca()  # 'obtener el eje actual'
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    #Incluir anotaci√≥n en la gr√°fica
    plt.annotate(r'('+str(x0)+','+str(f(x0))+')',xy=(x0, f(x0)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
        #Incluir anotaci√≥n en la gr√°fica
    plt.annotate(r'('+str(x0+h)+','+str(f(x0+h))+')',xy=(x0+h, f(x0+h)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    #Se muestra la grafica.
    plt.show()

print('\033[1m'+'SOLUCI√ìN:RECTA SECANTE\n'+'\033[0m')
#Llamada a la funci√≥n Graficasecante para pintar la funci√≥n y la secante
Graficasecante(x0,h)
```

### Explicaci√≥n de las secantes para cada valor de $h$

Para cada valor de $h$ la recta secante que pasa   $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$ tiene por ecuaci√≥n $y=\frac{f(x_0+h)-f(x_0)}{h}(x-x_0)+f(x_0)$, cuya pendiente es $\frac{f(x_0+h)-f(x_0)}{h}$.
Las gr√°ficas anteriores muestran que conforme $h$ toma valores m√°s peque√±os, la recta secante se aproxima a la recta tangente a la gr√°fica en el punto $(x_0,f(x_0))$.
La pregunta que surge es ¬øCu√°l es la pendiente de la recta tangente?

## Definici√≥n de la derivada de una funci√≥n $f$ en un punto $x_0$

La derivada de la funci√≥n $f(x)$ en un punto abscisa $x_0$ se escribe con el s√≠mbolo $f'(x)$ y se define como

$$f'(x_0)=lim_{h \to 0}\frac{f(x_0+h)-f(x_0)}{h}$$

a) Si el l√≠mite existe entonces diremos que la funci√≥n $f$ es derivable en el punto $x_0$.

b) La derivada de una funci√≥n en un punto es un n√∫mero real, que puede ser positivo, negativo o cero.

c) La derivada de una funci√≥n en un punto coincide con la pendiente de la recta tangente en ese punto

$\large{\textbf{Ejemplos  de funciones con sus tangentes en un punto }}$

Probamos con algunas funciones y sus rectas tangentes mediante la derivada en un punto

i) $f(x)=3x^2+5, \qquad x_0=-5$
La derivada para esta funci√≥n en $x_0=-5$, seg√∫n la definici√≥n es 

$$f'(-5)=lim_{h \to 0}\frac{f(-5+h)-f(-5)}{h}=lim_{h \to 0}\frac{3(-5+h)^2+5-(3(-5)^2+5)}{h}=-30$$

Teniendo en cuenta del apartado $c)$ de la definici√≥n se tiene que la pendiente de la recta tangente a la gr√°fica de la ecuaci√≥n en el punto $(-5,f(-5))=(-5,80)$ es $-30$. Por lo tanto la recta tangente en el punto $(-5,80)$ tiene por ecuaci√≥n $y=-30(x+5)+80$

### M√≥dulo 3: Grafica funci√≥n f dada y tangente que pasa por  $(ùë•_0,ùëì(ùë•_0))$  para $x_0$ y pendiente $m$ dados

```python
# ***************** CONSIDERACIONES PARA LA DEFINICI√ìN DE FUNCIONES*********************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x), la funcion arcoseno(x) como arcsin(x).
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN"
# Asingar valores para x0 y h despu√©s de la l√≠nea "AQU√ç SE ASIGNAN LOS VALORES x0 y m "
# ***************************************************************************************************

# *******************DEFINICI√ìN DE LA FUNCI√ìN********************************************************
def f(x):
    #Llamada de funciones de numpy
    from numpy import sin, cos, tan, arcsin, arccos, arctan, log
    
    # **************************************************************************** 
    #            AQU√ç SE DEFINE LA FUNCI√ìN    
    f=3*x**2+5
    
    # ****************************************************************************
    
    return f

# ********************************************************************************
#                 AQU√ç SE ASIGNAN LOS VALORES x0 Y m
x0=-5
m=-30

# ********************************************************************************


%matplotlib inline
def Graficatangente(x0,h):
    #Se importa numpy como np
    import numpy as np
    #De matplotlib se importa pyplot como plt
    import matplotlib.pyplot as plt

    # Se evalua la funci√≥n en el rango de [x0-h,x0+2h]
    x = np.linspace(x0-h,x0+2*h,101)
    #Se le pasa a la funcion los valores definidos en x.
    y = f(x)

    #Se calcula la recta tangente, de la forma y=m*(x-x0)+y0. Donde m es la pendiente 
    ytangente = m*(x-x0)+f(x0)  
    
    # Se grafica la funcion y la secante.
    plt.plot(x,y,'b',x0,f(x0),'om',x,ytangente,'r')
    
    # Indicar s√≥lo los valores requeridos en los ejes
    plt.xticks([x0])
    plt.yticks([f(x0),max(y)])
    # Se mueven los ejes coordenados con intersecci√≥n en el origen
    ax = plt.gca()  # 'obtener el eje actual'
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    #Incluir anotaci√≥n en la gr√°fica
    plt.annotate(r'('+str(x0)+','+str(f(x0))+')',xy=(x0, f(x0)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    plt.title('Grafica de la funci√≥n f y la recta en (x0,f(x0)) y pendiente m='+str(m))
    
    #Se muestra la grafica.
    plt.show()

#Llamada a la funci√≥n Graficasecante para pintarlas
print('\033[1m'+'SOLUCI√ìN: RECTA QUE PASA POR UN PUNTO Y PENDIENTE DADOS\n'+'\033[0m')
Graficatangente(x0,m)

```

# Parte II: FUNCI√ìN DERIVADA

Dada una funci√≥n $f: I \rightarrow \mathbb{R}$, la derivada de $f$ en x est√° definida por $f'(x)=\lim_{h \to 0}\frac{f(x+h)-f(x)}{f}$ siempre que exista el l√≠mite. Para todos los valores $x$ para los que exista el l√≠mite, $f'$ es una funci√≥n de $x$.

$\large{\textbf{Ejemplo}}$

Sea $f(x)=4x^2-3$, calcula la funci√≥n derivada $f'(x)$.

Por la definici√≥n se tiene

$$f'(x)=\lim_{h \to 0} \frac{4(x+h)^2-3-(4x^2-3)}{h}=\lim_{h \to 0}\frac{8xh+h^2}{h}=8x$$

En el M√≥culo 4 se halla y grafica la funcion derivada para una funci√≥n $f$ dada. Se grafica la funci√≥n inicial en color azul y la funci√≥n derivada en color rojo.

Como ejemplo inicial se toma $f(x)=4x^2-3$. Pero se pueden definir otras funciones.

```python
# ***************** PARA LA DEFINICI√ìN DE FUNCIONES*************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La funci√≥n arcoscosno(x) como acos(x), la fun√≠on arcotangente(x) como atan(x)
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN"
# **************************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp

# ****************************************************************************
import sympy as sp
x=sp.Symbol('x')
# **************************************************************************** 
#            AQU√ç SE DEFINE LA FUNCI√ìN
f=4*x**2-3

# ****************************************************************************

%matplotlib inline
from sympy.plotting import plot

fp=sp.diff(f,x)
p=plot(f,title='Gr√°fica de la funci√≥n f y de la funci√≥n derivada',line_color='blue',show=False)
p1=plot(fp,line_color='red',show=False)
p.append(p1[0])

print('\033[1m'+'SOLUCI√ìN: FUNCI√ìN DERIVADA\n'+'\033[0m')
print('Funci√≥n dada, en color azul, f(x)=',str(f)+'\n')
print('Funci√≥n derivada, en color rojo, f¬¥(x)=',str(fp)+'\n')

p.show()


```

## Algunas reglas de derivaci√≥n

$f(x)=C, \qquad f'(x)=0$

$f(x)=x, \qquad f'(x)=1$

$f(x)=x^n,  \  \  n \in \mathbb{R}, \qquad f'(x)=nx^{n-1}$

$f(x)=ln (x), \qquad f'(x)=\frac{1}{x}$

$f(x)=e^x, \qquad f'(x)=e^x$

$f(x)=a^x, \ \ a>0, \qquad f'(x)=a^x ln (a)$

$f(x)=sen(x), \qquad f'(x)=cos(x)$

$f(x)=cos(x), \qquad f'(x)=-sen(x)$

$f(x)=tan(x), \qquad f'(x)=1+ tan^2(x)$

$f(x)=arcsen(x), \qquad f'(x)=\frac{1}{\sqrt{1-x^2}}$

$f(x)=arccos(x), \qquad f'(x)=\frac{-1}{\sqrt{1-x^2}}$

## Propiedades de la derivada

Si $f$ y $g$ son funciones derivables en $x$, tambi√©n lo son $cf$, $f+g$, $fg$, $\frac{f}{g}$ y $f \circ g$, adem√°s se verifica

$[cf(x)]'=cf'(x),$

$[f(x)+g(x)]' = f'(x)+g'(x),$

$(f(x)g(x))' = f'(x)g(x)+f(x)g'(x)$

$(\frac{f}{g})'(x) = \frac{f'(x)g(x)-f(x)g'(x)}{(g(x))^2}, \qquad si \ \ \ g(x) \neq 0$

$(g \circ f)'(x) = g'(f(x))f'(x)$

$\large{\textbf{Ejemplo}}$

Dada la funci√≥n $f(x)=2x^3+1$, calcular la funci√≥n derivada y la derivada en el punto $x_0=3$, dibujar la gr√°fica de la funci√≥n $f$ y la recta tangente en el punto $(x_0,f(x_0))$

Aplicando las reglas de derivaci√≥n se tiene $f'(x)= 6x^2$, por otro lado, evaluando la funci√≥n $f$, y $f'$ en $x_0=3$ se tiene $f(3)=55$ y $f'(3)=54$, por lo que la recta tangente tiene por ecuaci√≥n $y=54(x-3)+55$

## M√ÅXIMOS Y M√çNIMOS RELATIVOS

$\textbf{Definici√≥n.}$

Una funci√≥n $f(x)$ tiene un m√°ximo relativo en $x=x_0$ si existe un intervalo abierto $(a,b)$ tal que $x_0\in (a,b)$ y $f(x) \geq f(x_0)$ para todo $x\in (a,b)$; $x \neq x_0$

An√°logamente, una funci√≥n $f(x)$ tiene un m√≠nimo relativo en $x=x_0$ si existe un intervalo abierto $(a,b)$ tal que $x_0\in (a,b)$ y $f(x) \leq f(x_0)$ para todo $x\in (a,b)$; $x \neq x_0$

A un m√°ximo o m√≠nimo relativo se le llama extremo relativo.

### Caracterizaci√≥n de los m√°ximos y m√≠nimos por la primera derivada

Si el punto $(x_0,f(x_0))$ es un punto m√°ximo o m√≠nimo relativo de de $f(x)$, entonces $f'(x_0)=0$, o $f(x)$ no es derivable en $x_0$

### Pasos para hallar los extremos relativos y clasificarlos

1. Calcular la primera derivada, $f'(x)$

2. Resolver la ecuaci√≥n $f'(x)=0$

3. Hallar la segunda derivada, $f''(x)$

4. Sustituir en $f''(x)$ las ra√≠ces $x$ obtenidas en el paso 2

$\qquad$ * Si $f''(x)<0$, entonces $f$ tiene un m√°ximo relativo en este punto.

$\qquad$ * Si $f''(x)>0$, entonces $f$ tiene un m√≠nimo relativo en este punto.

$\large{\textbf{Ejemplo}}$

Hallar los extremos relativos para la funcion $f(x)=2x^3-5x^2-1$.

Paso 1. Hallamos la primera derivada $f'(x)=6x^2-10x$

Paso 2. Resolvemos la ecuaci√≥n $f'(x)=6x^2-10x=0$, que da dos soluciones $\{0, \frac{5}{3}  \} $

Paso 3. Hallamos la segunda derivada $f''(x)=12x-10$

Paso 4. Sustituimos las ra√≠ces obtenidas en el Paso 2 en la segunda derivada $f''(x)$

* $f''(0)=-10$, al ser menor que cero estamos ante un m√°ximo relativo

* $f''(5/3)=10$, al ser mayor que cero estamos ante un m√≠nimo relativo

Por lo tanto la funci√≥n $f$ alcanza un m√°ximo relativo en $x=0$, donde $f(0)=-1$ y alcanza un m√≠nimo relativo en $x=\frac{5}{3}$, donde $f(\frac{5}{3})=-\frac{152}{27}$

```python
# ***************** PARA LA DEFINICI√ìN DE FUNCIONES*************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La funci√≥n arcoscosno(x) como acos(x), la fun√≠on arcotangente(x) como atan(x)
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f y asignar el valor para n despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN Y SE
# ASIGNA EL VALOR DE n"
# **************************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp
import sympy as sp
x=sp.Symbol('x')

# ****************************************************************************
#            AQU√ç SE DEFINE LA FUNCI√ìN Y SE ASIGNA EL VALOR DE n
f=2*x**3-5*x**2-1

# ****************************************************************************

%matplotlib inline
from sympy.plotting import plot


fp=sp.diff(f,x,1) #C√°lculo de la derivada como funci√≥n

sol=sp.solveset(fp, x, domain=sp.S.Reals) # Halla las ra√≠ces de f'(x)=0

sol=list(sol) # Convierte las soluciones de formato FiniteSet a formato Lista
sol=sorted(sol)

print('\033[1m'+'SOLUCI√ìN: C√ÅLCULO DE EXTREMOS RELATIVOS\n'+'\033[0m')
print('Funci√≥n inicial f(x)= '+str(f)+'\n')
print('Ecuaci√≥n f¬¥(x)=0 ===> '+ str(fp)+'=0\n')
print('Ra√≠ces de la ecuaci√≥n:'+str(sol)+'\n')

if len(sol)>0:
    fpp=sp.diff(f,x,2) #C√°lculo de la derivada como funci√≥n    
    print('Funci√≥n segunda derivada: f¬¥¬¥(x)='+str(fpp)+'\n')
    for i in range(0,len(sol)):
        fppx=fpp.subs(x,sol[i]) #Evaluar la segunda derivada en cada soluci√≥n        
        if fppx==0:
            print('En x='+str(sol[i])+', no se puede decir nada, pues en este valor la segunda derivada es cero')
        if fppx>0:
            fx=f.subs(x,sol[i]) #Evaluar f en la raiz
            print('En x='+str(sol[i])+', f¬¥¬¥('+str(sol[i])+')='+str(fppx) +', la funci√≥n alcanza un m√≠nimo relativo, y f('+str(sol[i])+')='+str(fx)+'\n') 
        if fppx<0:
            fx=f.subs(x,sol[i]) #Evaluar f en la raiz            
            print('En x='+str(sol[i])+', f¬¥¬¥('+str(sol[i])+')='+str(fppx) +', la funci√≥n alcanza un m√°ximo relativo, y f('+str(sol[i])+')='+str(fx)+'\n') 
    p=plot(f,(x,min(sol)-2,max(sol)+2),title='Grafica de la funci√≥n f',line_color='blue',show=False)
else:
    print('No existen extremos relativos')
    p=plot(f,title='Grafica de la funci√≥n f',line_color='blue',show=False)
     
p.show()
```

## MONOTON√çA

$\textbf{Definici√≥n.}$

Una funci√≥n $f$ es creciente en un intervalo $(a,b)$ si para cualquier par de n√∫meros $x_1,x_2 \in (a,b)$, $x_1 < x_2$ implica que $f(x_1)<f(x_2)$.

An√°logamente una funci√≥n $f$ es decreciente en un intervalo $(a,b)$ si para cualquier par de n√∫meros $x_1,x_2 \in (a,b)$, $x_1 < x_2$ implica que $f(x_1)>f(x_2)$.


### Criterios para funciones crecientes o decrecientes

Sea $f(x)$ una funci√≥n continua en $[a,b]$ y derivable en $(a,b)$

1. Si $f'(x)>0$  para todo $x \in (a,b)$, $f(x)$ es creciente en $[a,b]$

2. Si $f'(x)<0$  para todo $x \in (a,b)$, $f(x)$ es decreciente en $[a,b]$

3. Si $f'(x)=0$  para todo $x \in (a,b)$, $f(x)$ es constante en $[a,b]$

Es decir estudiar la monoton√≠a de una funci√≥n es estudiar los intervalos en los que la funci√≥n es creciente o decreciente. Estos intervalos est√°n separados por los extremos relativos y las discontinuidades de la primera derivada.


### Pasos para identificar los intervalos de crecimiento o decrecimiento

1. Se calculan los m√°ximos y m√≠nimos.

2. Se hallan las discontinuidades de la primera derivada de la funci√≥n $f'(x)$

3. Con los puntos de los pasos 1 y 2 se determinan los intervalos a analizar.

3. Se prueba un punto de cada uno de los intervalos determinados en el paso 3 y se clasifican los intervalos

$\qquad$ * Es intervalo de crecimiento si $f'(x)>0$

$\qquad$ * Es intervalo de decrecimiento si $f'(x)<0$

$\large{\textbf{Ejemplo}}$

Determina los interevalos de crecimiento y decrecimiento de la funci√≥n $f(x)=2x^3-5x^2-1$

Paso 1. Se calculan los valores donde $f$ alcanza m√°ximos y m√≠nimos. Para la funci√≥n dada, estos son $\{0, \frac{5}{3}  \} $

Paso 2. Se hallan discontinuidades de $f'(x)=6x^2-10x$. No existen

Paso 3. Con los puntos encontrados en los pasos 1 y 2, se determinan los intervalos a analizar:

* Intervalo $I_1=(-\infty,0)$, $f'(-1)=16>0$, por lo que es un intervalo de crecimiento

* Intervalo $I_2=(0,\frac{5}{3})$, $f'(1)=-4 < 0$, por lo que es un intervalo de decrecimiento

* Intervalo $I_3=(\frac{5}{3},\infty)$, $f'(2)=4>0$, por lo que es un intervlao de crecimiento

## CONCAVIDAD, CONVEXIDAD Y PUNTOS DE INFLEXI√ìN

$\textbf{Definici√≥n.}$

Sea $f(x)$ una funci√≥n derivable en $(a,b)$. La gr√°fica de $f(x)$ es convexa $(\cup)$ en $(a,b)$ si $f'(x)$ es creciente en $(a,b)$, y es c√≥ncava $(\cap)$ si $f'(x)$ es decreciente en $(a,b)$.

Un punto de inflexi√≥n de una funci√≥n es un punto en el que la funci√≥n cambia de c√≥ncava a convexa, o viceversa, y la tangente atraviesa la gr√°fica.


### Pasos para identificar los puntos de inflexi√≥n

1. Se calcula la segunda derivada, $f''(x)$

2. Se resuelve la ecuaci√≥n, $f''(x)=0$

3. Se halla la tercera derivada, $f'''(x)$

4. Se sustituyen las ra√≠ces halladas en el Paso 2 en la tercera derivada. Si $f'''(x) \neq 0$, son puntos de inflexi√≥n.

$\large{\textbf{Ejemplo}}$

Dada la funci√≥n $f(x)=2x^3-5x^2-1$, hallar los puntos de inflexi√≥n

Paso 1. Calculamos la segunda derivada, $f''(x)=12x-5$

Paso 2. Resolvemos la ecuaci√≥n $f''(x)=12x-5=0$, que tiene como soluci√≥n $x=\frac{5}{6}$

Paso 3. Hallamos la tercera derivada, $f'''(x)=12$

Paso 4. Sustituimos en la tercera derivada la ra√≠z obtenida en el Paso 2. $f'''(\frac{5}{6})=12$, que es distinto de cero, por lo que el punto $x=\frac{5}{6}$ es un punto de inflexi√≥n.

En el m√≥dulo 9, para una funci√≥n $f$ dada (no peri√≥dica), se calculan sus puntos de inflexi√≥n.

### M√≥dulo 9. Calcula los puntos de inflexi√≥n de una funci√≥n $f$ dada

```python
# ***************** PARA LA DEFINICI√ìN DE FUNCIONES*************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La funci√≥n arcoscosno(x) como acos(x), la fun√≠on arcotangente(x) como atan(x)
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f y asignar el valor para n despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN Y
# SE ASIGNA EL VALOR DE n"
# ************************************************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp
import sympy as sp
x=sp.Symbol('x')

# ****************************************************************************
#            AQU√ç SE DEFINE LA FUNCI√ìN Y SE ASIGNA EL VALOR DE n
f=2*x**3-5*x**2-1

# ****************************************************************************


%matplotlib inline
from sympy.plotting import plot

fpp=sp.diff(f,x,2) #C√°lculo de la derivada como funci√≥n

sol=sp.solveset(fpp, x, domain=sp.S.Reals) # Halla las ra√≠ces de f''(x)=0

sol=list(sol) # Convierte las soluciones de formato FiniteSet a formato Lista

sol=sorted(sol)


fppp=sp.diff(f,x,3) #C√°lculo de la derivada como funci√≥n

print('\033[1m'+'SOLUCI√ìN: PUNTOS DE INFLEXI√ìN\n'+'\033[0m')
if len(sol)==0:
    print('La funci√≥n f(x)='+str(f)+' no tiene puntos de inflexi√≥n\n')
else:
    for i in range(0,len(sol)):
        if fppp.subs(x,sol[i])!=0:
            print('('+str(sol[i])+','+str(f.subs(x,sol[i]))+')'+' es punto de inflexi√≥n de la funci√≥n f(x)='+str(f)+'\n')
        else:
            print('La funci√≥n f(x)='+str(f)+' no tiene puntos de inflexi√≥n\n')
            
if len(sol)!=0:
    p=plot(f,(x,min(sol)-2,max(sol)+2),title='Grafica de la funci√≥n f\n',line_color='blue',show=False)
else:
    p=plot(f,title='Grafica de la funci√≥n f',line_color='blue',show=False)    
p.show()              
                
```

### Criterio para el estudio de la curvatura

Sea $f(x)$ una funci√≥n cuya segunda derivada existe en $(a,b)$:
    
a) Si $f''(x)>0$ para todo $x \in (a,b)$, la gr√°fica de $f(x)$ es convexa en $(a,b)$

b) Si $f''(x)<0$ para todo $x \in (a,b)$, la gr√°fica de $f(x)$ es c√≥ncaba en $(a,b)$

Analizar la curvatura de una funci√≥n consiste en estudiar en qu√© intervalos es c√≥ncava o convexa. Cabe notar que los intervalos de curvatura est√° separados por los puntos de inflexi√≥n y las discontinuidades de $f''(x)$.

### Pasos para analizar la curvatura de una funci√≥n

1. Se calculan los puntos de inflexi√≥n

2. Se hallan las discontinuidades de la tercera derivada de la funci√≥n, $f'''(x)$

3. Con los puntos de los pasos 1 y 2 se determinan los intervalos a analizar.

4. Para cada intervalo determinado en el paso 3 se toma un punto del intervalo y se eval√∫a la segunda derivada $f''(x)$

$\qquad$ * Si $f''(x)>0$ la funci√≥n es convexa

$\qquad$ * Si $f''(x)<0$ la funci√≥n es c√≥ncava


$\large{\textbf{Ejemplo}}$

Dada la funci√≥n $f(x)=2x^3-5x^2-1$, analizar la curvatura

Paso 1. Calculamos los puntos de inflexi√≥n, tiene un punto de inflexi√≥n en $x=\frac{5}{6}$

Paso 2. Se hallan las discontinuidades de la tercera derivada de la funci√≥n, $f'''(x)=12$. No tiene

Paso 3. Se determinan los intervalos a analizar $(-\infty,\frac{5}{6})$ y $(\frac{5}{6},+\infty)$

Paso 4. Identificamos la curvatura en cada intervalo
* Calculamos la segunda derivada, $f''(x)=12x-10$

* En el intervalo $I_1=(-\infty,\frac{5}{6})$, $f''(0)=-10$ que es menor que cero, por lo que es c√≥ncava.

* En el intervalo $I_2=(-\infty,\frac{5}{6})$, $f''(1)=2$ que es mayor que cero, por lo que es convexa.


En el m√≥dulo 10, para una funci√≥n  ùëì  dada (no peri√≥dica), se identifican los intervalos de curvatura.

### M√≥dulo 10. Identifica los intervalos de curvatura de una funci√≥n $f$ dada

```python
# ***************** PARA LA DEFINICI√ìN DE FUNCIONES************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La funci√≥n seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La funci√≥n arcoscosno(x) como acos(x), la fun√≠on arcotangente(x) como atan(x)
# La funci√≥n logaritmo neperiano como log(x). La funci√≥n exponencial como exp(x).
# La funci√≥n valor absoluto como abs(x).
# Definir la funci√≥n f y asignar el valor para n despu√©s de la l√≠nea "AQU√ç SE DEFINE LA FUNCI√ìN Y
# SE ASIGNA EL VALOR DE n"
# ***********************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp
import sympy as sp
x=sp.Symbol('x')

# ****************************************************************************
#            AQU√ç SE DEFINE LA FUNCI√ìN Y SE ASIGNA EL VALOR DE n
f=2*x**3-5*x**2-1
#f=5+(x**2-2)
# ****************************************************************************


%matplotlib inline
from sympy.plotting import plot


fpp=sp.diff(f,x,2) #C√°lculo de la segunda derivada como funci√≥n

# Halla los puntos de inflexi√≥n

sol=sp.solveset(fpp, x, domain=sp.S.Reals) # Halla las ra√≠ces de f''(x)=0

sol=list(sol) # Convierte las soluciones de formato FiniteSet a formato Lista

sol=sorted(sol) #Ordenamos la lista de menor a mayor

fppp=sp.diff(f,x,3) #C√°lculo de la derivada como funci√≥n

print('\033[1m'+'SOLUCI√ìN: AN√ÅLISIS DE CURVATURA\n'+'\033[0m')
print('Funci√≥n dada: f(x)='+str(f)+'\n')
print('Segunda derivada: f¬¥¬¥(x)='+str(fpp)+'\n')


solinflex=list()
if len(sol)==0:
    print('La funci√≥n f(x)='+str(f)+' no tiene puntos de inflexi√≥n\n')
else:
    for i in range(0,len(sol)):
        if fppp.subs(x,sol[i])!=0:           
            solinflex.append(sol[i])            
            print('('+str(sol[i])+','+str(f.subs(x,sol[i]))+')'+' es punto de inflexi√≥n\n')
        #else:
            #print('La funci√≥n f(x)='+str(f)+' no tiene puntos de inflexi√≥n\n')

#Puntos de inflexi√≥n en solinflex[]
# Se identifica la curvatura

if len(solinflex)==0:    # Se definen los intervalos
    solinflex[len(solinflex):]=[0,1]
else:
    solinflex[len(solinflex):]=[min(solinflex)-1,max(solinflex)+1]

solinflex=sorted(solinflex)    
    
linter=list()
for i in range(0,len(solinflex)): # Se define la lista linter con los extremos de los intervalos
    linter.append(str(solinflex[i]))

linter[0]=('-‚àû'); linter[len(linter)-1]=('+‚àû')


#An√°lisis de intervalos de curvatura

for i in range(0,len(solinflex)-1):
    xm=(solinflex[i+1]+solinflex[i])/2
    fppval=fpp.subs(x,xm)
    if fppval==0:
        fcurva='constante'
    else:
        if fppval>0:
            fcurva='convexa'
        else:
            fcurva='c√≥ncava'
    print('En el intervalo ('+linter[i]+','+linter[i+1]+') la funci√≥n es '+fcurva+'\n')


p=plot(f,(x,min(solinflex)-1,max(solinflex)+1),title='Grafica de la funci√≥n f',line_color='blue',show=False)
p.show()    
```
