## Aproximaciones mediante secantes

Dada una función $f(x): I \rightarrow \mathbb{R}$, un valor $x_0 \in I$ y un valor $h$ trazamos la recta secante que pasa por los puntos $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$. Conforme el valor de $h$ se va haciendo más pequeño las rectas secantes se irán aproximando a la recta tangente a la gráfica de la función en el punto $(x_0,f(x_0))$. 

$\large{\textbf{Ejemplos}}$

$\large{\textbf{Para un valor de $\textit{h}$ determinado}}$

Sea la función $f(x)=3x^2+5$, graficar la recta secante, que pasa por los puntos $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$ para los valores $x_0$ y $h$ que se asignen, como ejemplo inicial $x_0=-5$ y $h=8$.

## Módulo 1: Grafica función f dada y secante que pasa por $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$, para $x_0$ y $h$ dados

```python
# ***************** CONSIDERACIONES PARA LA DEFINICIÓN DE FUNCIONES*********************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La función seno(x) se define como sin(x), la funcion arcoseno(x) como arcsin(x).
# La función logaritmo neperiano como log(x). La función exponencial como exp(x).
# La función valor absoluto como abs(x).
# Definir la función f después de la línea "AQUÍ SE DEFINE LA FUNCIÓN"
# Asingar valores para x0 y h después de la línea "AQUÍ SE ASIGNAN LOS VALORES x0 y h "
# ***************************************************************************************************

# *******************DEFINICIÓN DE LA FUNCIÓN********************************************************
def f(x):
    #Llamada de funciones de numpy
    from numpy import sin, cos, tan, arcsin, arccos, arctan, log, exp
    
    # **************************************************************************** 
    #            AQUÍ SE DEFINE LA FUNCIÓN    
    f=3*x**2+5
    
    # ****************************************************************************
    
    return f

# ********************************************************************************
#                 AQUÍ SE ASIGNAN LOS VALORES x0 Y h 
x0=-5
h=8

# ********************************************************************************


%matplotlib inline
def Graficasecante(x0,h):
    #De matplotlib se importa pyplot como plt
    import matplotlib.pyplot as plt
    #Se importa numpy como np
    import numpy as np

    
    # Se evalua la función en el rango de [x0-h,x0+2h]
    x = np.linspace(x0-h,x0+2*h,101)
    #Se le pasa a la funcion los valores definidos en x.
    y = f(x)

    #Se calcula la recta secante, de la forma y=(y1-y0)/(x1-x0)*(x-x0)+y0
    
    rsec = (f(x0+h)-f(x0))/h*(x-x0)+f(x0)  
    
    #Para el valor h se grafica la recta secante que pasa por (x0,f(x0)) y (x0+h,f(x0+h))
    
    # Se grafica la funcion y la secante.
    plt.title('Secante con h  = '+str(h))
    plt.plot(x,y,'b',x0,f(x0),'om',x0+h,f(x0+h),'om',x,rsec,'r')
    
    # Indicar sólo los valores requeridos en los ejes
    plt.xticks([x0,x0+h])
    plt.yticks([f(x0),f(x0+h),max(y)])
    # Se mueven los ejes coordenados con intersección en el origen
    ax = plt.gca()  # 'obtener el eje actual'
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    #Incluir anotación en la gráfica
    plt.annotate(r'('+str(x0)+','+str(f(x0))+')',xy=(x0, f(x0)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
        #Incluir anotación en la gráfica
    plt.annotate(r'('+str(x0+h)+','+str(f(x0+h))+')',xy=(x0+h, f(x0+h)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    #Se muestra la grafica.
    plt.show()

print('\033[1m'+'SOLUCIÓN:RECTA SECANTE\n'+'\033[0m')
#Llamada a la función Graficasecante para pintar la función y la secante
Graficasecante(x0,h)
```

### Explicación de las secantes para cada valor de $h$

Para cada valor de $h$ la recta secante que pasa   $(x_0,f(x_0))$ y $(x_0+h,f(x_0+h))$ tiene por ecuación $y=\frac{f(x_0+h)-f(x_0)}{h}(x-x_0)+f(x_0)$, cuya pendiente es $\frac{f(x_0+h)-f(x_0)}{h}$.
Las gráficas anteriores muestran que conforme $h$ toma valores más pequeños, la recta secante se aproxima a la recta tangente a la gráfica en el punto $(x_0,f(x_0))$.
La pregunta que surge es ¿Cuál es la pendiente de la recta tangente?

## Definición de la derivada de una función $f$ en un punto $x_0$

La derivada de la función $f(x)$ en un punto abscisa $x_0$ se escribe con el símbolo $f'(x)$ y se define como

$$f'(x_0)=lim_{h \to 0}\frac{f(x_0+h)-f(x_0)}{h}$$

a) Si el límite existe entonces diremos que la función $f$ es derivable en el punto $x_0$.

b) La derivada de una función en un punto es un número real, que puede ser positivo, negativo o cero.

c) La derivada de una función en un punto coincide con la pendiente de la recta tangente en ese punto

$\large{\textbf{Ejemplos  de funciones con sus tangentes en un punto }}$

Probamos con algunas funciones y sus rectas tangentes mediante la derivada en un punto

i) $f(x)=3x^2+5, \qquad x_0=-5$
La derivada para esta función en $x_0=-5$, según la definición es 

$$f'(-5)=lim_{h \to 0}\frac{f(-5+h)-f(-5)}{h}=lim_{h \to 0}\frac{3(-5+h)^2+5-(3(-5)^2+5)}{h}=-30$$

Teniendo en cuenta del apartado $c)$ de la definición se tiene que la pendiente de la recta tangente a la gráfica de la ecuación en el punto $(-5,f(-5))=(-5,80)$ es $-30$. Por lo tanto la recta tangente en el punto $(-5,80)$ tiene por ecuación $y=-30(x+5)+80$

### Módulo 3: Grafica función f dada y tangente que pasa por  $(𝑥_0,𝑓(𝑥_0))$  para $x_0$ y pendiente $m$ dados

```python
# ***************** CONSIDERACIONES PARA LA DEFINICIÓN DE FUNCIONES*********************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La función seno(x) se define como sin(x), la funcion arcoseno(x) como arcsin(x).
# La función logaritmo neperiano como log(x). La función exponencial como exp(x).
# La función valor absoluto como abs(x).
# Definir la función f después de la línea "AQUÍ SE DEFINE LA FUNCIÓN"
# Asingar valores para x0 y h después de la línea "AQUÍ SE ASIGNAN LOS VALORES x0 y m "
# ***************************************************************************************************

# *******************DEFINICIÓN DE LA FUNCIÓN********************************************************
def f(x):
    #Llamada de funciones de numpy
    from numpy import sin, cos, tan, arcsin, arccos, arctan, log
    
    # **************************************************************************** 
    #            AQUÍ SE DEFINE LA FUNCIÓN    
    f=3*x**2+5
    
    # ****************************************************************************
    
    return f

# ********************************************************************************
#                 AQUÍ SE ASIGNAN LOS VALORES x0 Y m
x0=-5
m=-30

# ********************************************************************************


%matplotlib inline
def Graficatangente(x0,h):
    #Se importa numpy como np
    import numpy as np
    #De matplotlib se importa pyplot como plt
    import matplotlib.pyplot as plt

    # Se evalua la función en el rango de [x0-h,x0+2h]
    x = np.linspace(x0-h,x0+2*h,101)
    #Se le pasa a la funcion los valores definidos en x.
    y = f(x)

    #Se calcula la recta tangente, de la forma y=m*(x-x0)+y0. Donde m es la pendiente 
    ytangente = m*(x-x0)+f(x0)  
    
    # Se grafica la funcion y la secante.
    plt.plot(x,y,'b',x0,f(x0),'om',x,ytangente,'r')
    
    # Indicar sólo los valores requeridos en los ejes
    plt.xticks([x0])
    plt.yticks([f(x0),max(y)])
    # Se mueven los ejes coordenados con intersección en el origen
    ax = plt.gca()  # 'obtener el eje actual'
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    #Incluir anotación en la gráfica
    plt.annotate(r'('+str(x0)+','+str(f(x0))+')',xy=(x0, f(x0)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    plt.title('Grafica de la función f y la recta en (x0,f(x0)) y pendiente m='+str(m))
    
    #Se muestra la grafica.
    plt.show()

#Llamada a la función Graficasecante para pintarlas
print('\033[1m'+'SOLUCIÓN: RECTA QUE PASA POR UN PUNTO Y PENDIENTE DADOS\n'+'\033[0m')
Graficatangente(x0,m)

```

# Parte II: FUNCIÓN DERIVADA

Dada una función $f: I \rightarrow \mathbb{R}$, la derivada de $f$ en x está definida por $f'(x)=\lim_{h \to 0}\frac{f(x+h)-f(x)}{f}$ siempre que exista el límite. Para todos los valores $x$ para los que exista el límite, $f'$ es una función de $x$.

$\large{\textbf{Ejemplo}}$

Sea $f(x)=4x^2-3$, calcula la función derivada $f'(x)$.

Por la definición se tiene

$$f'(x)=\lim_{h \to 0} \frac{4(x+h)^2-3-(4x^2-3)}{h}=\lim_{h \to 0}\frac{8xh+h^2}{h}=8x$$

En el Móculo 4 se halla y grafica la funcion derivada para una función $f$ dada. Se grafica la función inicial en color azul y la función derivada en color rojo.

Como ejemplo inicial se toma $f(x)=4x^2-3$. Pero se pueden definir otras funciones.

### Módulo 4: Calcula la función derivada de una función $f$ dada y muestra las gráficas de la función inicial y de la función derivada.

```python
# ***************** PARA LA DEFINICIÓN DE FUNCIONES*************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La función seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La función arcoscosno(x) como acos(x), la funíon arcotangente(x) como atan(x)
# La función logaritmo neperiano como log(x). La función exponencial como exp(x).
# La función valor absoluto como abs(x).
# Definir la función f después de la línea "AQUÍ SE DEFINE LA FUNCIÓN"
# **************************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp

# ****************************************************************************
import sympy as sp
x=sp.Symbol('x')
# **************************************************************************** 
#            AQUÍ SE DEFINE LA FUNCIÓN
f=4*x**2-3

# ****************************************************************************

%matplotlib inline
from sympy.plotting import plot

fp=sp.diff(f,x)
p=plot(f,title='Gráfica de la función f y de la función derivada',line_color='blue',show=False)
p1=plot(fp,line_color='red',show=False)
p.append(p1[0])

print('\033[1m'+'SOLUCIÓN: FUNCIÓN DERIVADA\n'+'\033[0m')
print('Función dada, en color azul, f(x)=',str(f)+'\n')
print('Función derivada, en color rojo, f´(x)=',str(fp)+'\n')

p.show()


```

## Algunas reglas de derivación

$f(x)=C, \qquad f'(x)=0$

$f(x)=x, \qquad f'(x)=1$

$f(x)=x^n,  \  \  n \in \mathbb{R}, \qquad f'(x)=nx^{n-1}$

$f(x)=ln (x), \qquad f'(x)=\frac{1}{x}$

$f(x)=e^x, \qquad f'(x)=e^x$

$f(x)=a^x, \ \ a>0, \qquad f'(x)=a^x ln (a)$

$f(x)=sen(x), \qquad f'(x)=cos(x)$

$f(x)=cos(x), \qquad f'(x)=-sen(x)$

$f(x)=tan(x), \qquad f'(x)=1+ tan^2(x)$

$f(x)=arcsen(x), \qquad f'(x)=\frac{1}{\sqrt{1-x^2}}$

$f(x)=arccos(x), \qquad f'(x)=\frac{-1}{\sqrt{1-x^2}}$

## Propiedades de la derivada

Si $f$ y $g$ son funciones derivables en $x$, también lo son $cf$, $f+g$, $fg$, $\frac{f}{g}$ y $f \circ g$, además se verifica

$[cf(x)]'=cf'(x),$

$[f(x)+g(x)]' = f'(x)+g'(x),$

$(f(x)g(x))' = f'(x)g(x)+f(x)g'(x)$

$(\frac{f}{g})'(x) = \frac{f'(x)g(x)-f(x)g'(x)}{(g(x))^2}, \qquad si \ \ \ g(x) \neq 0$

$(g \circ f)'(x) = g'(f(x))f'(x)$

Aplicando las reglas de derivación se tiene $f'(x)= 6x^2$, por otro lado, evaluando la función $f$, y $f'$ en $x_0=3$ se tiene $f(3)=55$ y $f'(3)=54$, por lo que la recta tangente tiene por ecuación $y=54(x-3)+55$

## MÁXIMOS Y MÍNIMOS RELATIVOS

$\textbf{Definición.}$

Una función $f(x)$ tiene un máximo relativo en $x=x_0$ si existe un intervalo abierto $(a,b)$ tal que $x_0\in (a,b)$ y $f(x) \geq f(x_0)$ para todo $x\in (a,b)$; $x \neq x_0$

Análogamente, una función $f(x)$ tiene un mínimo relativo en $x=x_0$ si existe un intervalo abierto $(a,b)$ tal que $x_0\in (a,b)$ y $f(x) \leq f(x_0)$ para todo $x\in (a,b)$; $x \neq x_0$

A un máximo o mínimo relativo se le llama extremo relativo.

### Caracterización de los máximos y mínimos por la primera derivada

Si el punto $(x_0,f(x_0))$ es un punto máximo o mínimo relativo de de $f(x)$, entonces $f'(x_0)=0$, o $f(x)$ no es derivable en $x_0$

### Pasos para hallar los extremos relativos y clasificarlos

1. Calcular la primera derivada, $f'(x)$

2. Resolver la ecuación $f'(x)=0$

3. Hallar la segunda derivada, $f''(x)$

4. Sustituir en $f''(x)$ las raíces $x$ obtenidas en el paso 2

$\qquad$ * Si $f''(x)<0$, entonces $f$ tiene un máximo relativo en este punto.

$\qquad$ * Si $f''(x)>0$, entonces $f$ tiene un mínimo relativo en este punto.

$\large{\textbf{Ejemplo}}$

Hallar los extremos relativos para la funcion $f(x)=2x^3-5x^2-1$.

Paso 1. Hallamos la primera derivada $f'(x)=6x^2-10x$

Paso 2. Resolvemos la ecuación $f'(x)=6x^2-10x=0$, que da dos soluciones $\{0, \frac{5}{3}  \} $

Paso 3. Hallamos la segunda derivada $f''(x)=12x-10$

Paso 4. Sustituimos las raíces obtenidas en el Paso 2 en la segunda derivada $f''(x)$

* $f''(0)=-10$, al ser menor que cero estamos ante un máximo relativo

* $f''(5/3)=10$, al ser mayor que cero estamos ante un mínimo relativo

Por lo tanto la función $f$ alcanza un máximo relativo en $x=0$, donde $f(0)=-1$ y alcanza un mínimo relativo en $x=\frac{5}{3}$, donde $f(\frac{5}{3})=-\frac{152}{27}$

## MONOTONÍA

$\textbf{Definición.}$

Una función $f$ es creciente en un intervalo $(a,b)$ si para cualquier par de números $x_1,x_2 \in (a,b)$, $x_1 < x_2$ implica que $f(x_1)<f(x_2)$.

Análogamente una función $f$ es decreciente en un intervalo $(a,b)$ si para cualquier par de números $x_1,x_2 \in (a,b)$, $x_1 < x_2$ implica que $f(x_1)>f(x_2)$.


### Criterios para funciones crecientes o decrecientes

Sea $f(x)$ una función continua en $[a,b]$ y derivable en $(a,b)$

1. Si $f'(x)>0$  para todo $x \in (a,b)$, $f(x)$ es creciente en $[a,b]$

2. Si $f'(x)<0$  para todo $x \in (a,b)$, $f(x)$ es decreciente en $[a,b]$

3. Si $f'(x)=0$  para todo $x \in (a,b)$, $f(x)$ es constante en $[a,b]$

Es decir estudiar la monotonía de una función es estudiar los intervalos en los que la función es creciente o decreciente. Estos intervalos están separados por los extremos relativos y las discontinuidades de la primera derivada.


### Pasos para identificar los intervalos de crecimiento o decrecimiento

1. Se calculan los máximos y mínimos.

2. Se hallan las discontinuidades de la primera derivada de la función $f'(x)$

3. Con los puntos de los pasos 1 y 2 se determinan los intervalos a analizar.

3. Se prueba un punto de cada uno de los intervalos determinados en el paso 3 y se clasifican los intervalos

$\qquad$ * Es intervalo de crecimiento si $f'(x)>0$

$\qquad$ * Es intervalo de decrecimiento si $f'(x)<0$

$\large{\textbf{Ejemplo}}$

Determina los interevalos de crecimiento y decrecimiento de la función $f(x)=2x^3-5x^2-1$

Paso 1. Se calculan los valores donde $f$ alcanza máximos y mínimos. Para la función dada, estos son $\{0, \frac{5}{3}  \} $

Paso 2. Se hallan discontinuidades de $f'(x)=6x^2-10x$. No existen

Paso 3. Con los puntos encontrados en los pasos 1 y 2, se determinan los intervalos a analizar:

* Intervalo $I_1=(-\infty,0)$, $f'(-1)=16>0$, por lo que es un intervalo de crecimiento

* Intervalo $I_2=(0,\frac{5}{3})$, $f'(1)=-4 < 0$, por lo que es un intervalo de decrecimiento

* Intervalo $I_3=(\frac{5}{3},\infty)$, $f'(2)=4>0$, por lo que es un intervlao de crecimiento

## CONCAVIDAD, CONVEXIDAD Y PUNTOS DE INFLEXIÓN

$\textbf{Definición.}$

Sea $f(x)$ una función derivable en $(a,b)$. La gráfica de $f(x)$ es convexa $(\cup)$ en $(a,b)$ si $f'(x)$ es creciente en $(a,b)$, y es cóncava $(\cap)$ si $f'(x)$ es decreciente en $(a,b)$.

Un punto de inflexión de una función es un punto en el que la función cambia de cóncava a convexa, o viceversa, y la tangente atraviesa la gráfica.


### Pasos para identificar los puntos de inflexión

1. Se calcula la segunda derivada, $f''(x)$

2. Se resuelve la ecuación, $f''(x)=0$

3. Se halla la tercera derivada, $f'''(x)$

4. Se sustituyen las raíces halladas en el Paso 2 en la tercera derivada. Si $f'''(x) \neq 0$, son puntos de inflexión.

$\large{\textbf{Ejemplo}}$

Dada la función $f(x)=2x^3-5x^2-1$, hallar los puntos de inflexión

Paso 1. Calculamos la segunda derivada, $f''(x)=12x-5$

Paso 2. Resolvemos la ecuación $f''(x)=12x-5=0$, que tiene como solución $x=\frac{5}{6}$

Paso 3. Hallamos la tercera derivada, $f'''(x)=12$

Paso 4. Sustituimos en la tercera derivada la raíz obtenida en el Paso 2. $f'''(\frac{5}{6})=12$, que es distinto de cero, por lo que el punto $x=\frac{5}{6}$ es un punto de inflexión.

### Criterio para el estudio de la curvatura

Sea $f(x)$ una función cuya segunda derivada existe en $(a,b)$:
    
a) Si $f''(x)>0$ para todo $x \in (a,b)$, la gráfica de $f(x)$ es convexa en $(a,b)$

b) Si $f''(x)<0$ para todo $x \in (a,b)$, la gráfica de $f(x)$ es cóncaba en $(a,b)$

Analizar la curvatura de una función consiste en estudiar en qué intervalos es cóncava o convexa. Cabe notar que los intervalos de curvatura está separados por los puntos de inflexión y las discontinuidades de $f''(x)$.

### Pasos para analizar la curvatura de una función

1. Se calculan los puntos de inflexión

2. Se hallan las discontinuidades de la tercera derivada de la función, $f'''(x)$

3. Con los puntos de los pasos 1 y 2 se determinan los intervalos a analizar.

4. Para cada intervalo determinado en el paso 3 se toma un punto del intervalo y se evalúa la segunda derivada $f''(x)$

$\qquad$ * Si $f''(x)<0$ la función es cóncava

En el módulo 10, para una función  𝑓  dada (no periódica), se identifican los intervalos de curvatura.

### Módulo 10. Identifica los intervalos de curvatura de una función $f$ dada

```python
# ***************** PARA LA DEFINICIÓN DE FUNCIONES************************************************
# El producto hay que indicarlo con asterisco (*).Las potencias se indican con doble asterisco (**).
# La función seno(x) se define como sin(x),# la funcion arcoseno(x) SE DEFINE COMO asin(x),
# La función arcoscosno(x) como acos(x), la funíon arcotangente(x) como atan(x)
# La función logaritmo neperiano como log(x). La función exponencial como exp(x).
# La función valor absoluto como abs(x).
# Definir la función f y asignar el valor para n después de la línea "AQUÍ SE DEFINE LA FUNCIÓN Y
# SE ASIGNA EL VALOR DE n"
# ***********************************************************************************************
#Llamada de funciones de sympy
from sympy import sin, cos, tan, asin, acos, atan, log, exp
import sympy as sp
x=sp.Symbol('x')

# ****************************************************************************
#            AQUÍ SE DEFINE LA FUNCIÓN Y SE ASIGNA EL VALOR DE n
f=2*x**3-5*x**2-1
#f=5+(x**2-2)
# ****************************************************************************


%matplotlib inline
from sympy.plotting import plot


fpp=sp.diff(f,x,2) #Cálculo de la segunda derivada como función

# Halla los puntos de inflexión

sol=sp.solveset(fpp, x, domain=sp.S.Reals) # Halla las raíces de f''(x)=0

sol=list(sol) # Convierte las soluciones de formato FiniteSet a formato Lista

sol=sorted(sol) #Ordenamos la lista de menor a mayor

fppp=sp.diff(f,x,3) #Cálculo de la derivada como función

print('\033[1m'+'SOLUCIÓN: ANÁLISIS DE CURVATURA\n'+'\033[0m')
print('Función dada: f(x)='+str(f)+'\n')
print('Segunda derivada: f´´(x)='+str(fpp)+'\n')


solinflex=list()
if len(sol)==0:
    print('La función f(x)='+str(f)+' no tiene puntos de inflexión\n')
else:
    for i in range(0,len(sol)):
        if fppp.subs(x,sol[i])!=0:           
            solinflex.append(sol[i])            
            print('('+str(sol[i])+','+str(f.subs(x,sol[i]))+')'+' es punto de inflexión\n')
        #else:
            #print('La función f(x)='+str(f)+' no tiene puntos de inflexión\n')

#Puntos de inflexión en solinflex[]
# Se identifica la curvatura

if len(solinflex)==0:    # Se definen los intervalos
    solinflex[len(solinflex):]=[0,1]
else:
    solinflex[len(solinflex):]=[min(solinflex)-1,max(solinflex)+1]

solinflex=sorted(solinflex)    
    
linter=list()
for i in range(0,len(solinflex)): # Se define la lista linter con los extremos de los intervalos
    linter.append(str(solinflex[i]))

linter[0]=('-∞'); linter[len(linter)-1]=('+∞')


#Análisis de intervalos de curvatura

for i in range(0,len(solinflex)-1):
    xm=(solinflex[i+1]+solinflex[i])/2
    fppval=fpp.subs(x,xm)
    if fppval==0:
        fcurva='constante'
    else:
        if fppval>0:
            fcurva='convexa'
        else:
            fcurva='cóncava'
    print('En el intervalo ('+linter[i]+','+linter[i+1]+') la función es '+fcurva+'\n')


p=plot(f,(x,min(solinflex)-1,max(solinflex)+1),title='Grafica de la función f',line_color='blue',show=False)
p.show()    
```
