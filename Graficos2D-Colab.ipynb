## **Gráficos de Polilíneas**

```
# Se importan las librerías necesarias
import numpy as np
from matplotlib import pyplot as plt
```

```
# Se definen los vectores
y = 10+np.random.random(20)
x = [x for x in range(len(y))]
# Mejor convertir la lista x en array
x=np.array(x)

# La gráfica
plt.plot(x,y,'-')
plt.title("Gráfico de Polilínea")
plt.grid(axis='both')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend(['Polilínea'])

plt.show()
```

```

# Se redefinen 'x' e 'y'. Aquí se usa random.randn en lugar de random.random
y = 200 + np.random.randn(100)
x = [x for x in range(len(y))]
# Graficamos
plt.plot(x,y,'-')
plt.fill_between(x, y, 195, where=(y > 195), facecolor='g', alpha=0.6)
plt.title("Ejemplo de gráfico de polilínea")
plt.show()
print(np.random.randn(10))
```

# **Gráficos de curvas**

En esta sección se verá diferentes opciones para graficar curvas 2D
La idea inicial es la misma que para una polilínea. Es decir, se define el vector de abscisas y el de ordenadas, y luego se grafican como polilíneas

## **Gráficos de curvas en coordenadas cartesianas**

```
# Se definen vectores
x=np.arange(-np.pi, np.pi, 0.05)
# Definimos la Función
#y=np.sin(x)
#y=2*x**3+5
y=np.exp(-x)-5

# Se grafican los ejes
xx=np.array([x[0],x[len(x)-1]])
xy=np.array([0,0])
yy=np.array([np.min(y), np.max(y)])
plt.plot(xx,xy,'b')
plt.plot(xy,yy,'b')

# Ampliar el área de dibujo
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(y.min() * 1.1, y.max() * 1.1)


#Gráfico de la curva
plt.plot(x,y,'r')
plt.show()
```

Gráfico de dos curvas y sombreado entre ellas

```
# Se definen nuevos vectores
x=np.linspace(-2*np.pi,2*np.pi,50)
y1=np.sin(x)
y2=np.cos(x)+2

# Gráfico de las curvas
plt.plot(x,y1, 'r')
plt.plot(x,y2,color='magenta', linewidth=2, linestyle="--")

# Sombreado el área entre las curvas
plt.fill_between(x, y1, y2, facecolor='g', alpha=0.6)

#Título, etiquetas a los ejes y leyenda
plt.legend(['Sine','Cosine'])
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Área sombreada')

# Gráfico de los ejes coordenados desde el origen
plt.arrow(0, 0, 1, 0, head_width=0.1, head_length=0.1, fc='k', ec='k')
plt.arrow(0, 0, 0, 1, head_width=0.1, head_length=0.1, fc='k', ec='k')

# Otra forma de dibujar los ejes, mediante anotaciones
#plt.annotate("", xy=(1, 0), xytext=(0, 0),arrowprops=dict(arrowstyle="->"))
#plt.annotate("", xy=(0, 1), xytext=(0, 0),arrowprops=dict(arrowstyle="->"))

# Se puede cambiar la escala del eje y, para que ver el gráfico proporcional
#plt.ylim(-4, 4)


plt.show()
```

Gráfica de curvas cambiando la presentación por defecto

```
# Se definen vectores
x=np.linspace(-np.pi,np.pi,50)
y1=np.sin(x)
y2=np.cos(x) #+2

# Gráfico de las curvas
plt.plot(x,y1, 'r')
plt.plot(x,y2,color='green', linewidth=1 )

#Título
plt.title('Funciones con anotaciones')

# Indicar sólo los valores requeridos en los ejes
plt.xticks([-np.pi, np.pi])
plt.yticks([-1, 0, 1])

# Se mueven los ejes coordenados con intersección en el origen
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


# Anotaciones
t = 2 * np.pi / 3
# Anotación en la función sin(x)
plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.0, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Anotación en la función cos(x)
plt.plot([t, t], [0, np.cos(t)], color='green', linewidth=2.0, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='green')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```

# Gráficos de contornos

Gráfico de contornos, o lo que se conoce como curvas de nivel de una superficie.
Hay más información en el cuaderno de Gráficos 3D

```
# Se define la función de la superficie
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

# Se define la malla de gráfico de la superficie
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

# Las zonas de nivel (8 niveles)
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
# Las curvas de nivel (8 niveles)
plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)

plt.show()
```

## **Graficos de curvas en coordenadas polares**

```
import numpy as np
import matplotlib.pyplot as plt

#Se crea el rango de valores angulares. linspace por defecto genera 50 valores
theta = np.linspace(0,2*np.pi)

#Se calcula el valor de 'r' en función del ángulo 'theta'
#r = 2*np.abs(np.cos(theta))
#r = 5 + 10*theta
a=5
r=a*(1+np.cos(theta))

#Se crea la grafica en coordenadas polares pasando 
#el angulo theta y los valores de r
plt.polar(theta, r)

# Se crea una grilla particular de los angulos 45,90 y 360
plt.thetagrids(range(45, 360, 90))
# Se crea una grilla particular en función del radio
plt.rgrids(np.arange(0, 2*a, 3), angle=0);

plt.title('Curva en coordenadas polares')

#Se genera la grafica
plt.show()
```

Dos curvas, en coordenadas polares, en dos ventanas distintas

```
import numpy as np
import matplotlib.pyplot as plt

theta = np.arange(0,2*np.pi,0.05)

fig = plt.figure()

r = 5 + 10*theta
ax = fig.add_subplot(211, projection="polar")
ax.plot(theta,r)

r = 2*np.abs(np.cos(theta))
ax = fig.add_subplot(212, projection="polar")
ax.plot(theta,r)

plt.show()
```

## **Gráficos de curvas en coordenadas paramétricas**
Se calcula los valores para cada coordenada en función del parámetro y luego se grafica como en coordenadas cartesianas

```
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,2*np.pi)
r=10
x=r*np.cos(t)
y=r*np.sin(t)
plt.plot(x,y)

# Establecer igual relación de forma de los ejes
plt.axis('equal')

plt.title('Igual relación de forma de los ejes')

plt.show
```
